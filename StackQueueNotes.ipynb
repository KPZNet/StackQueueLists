```python
from QueueRunner import QueueRunner
from Queues import QueueLinkedList
from Queues import QueueList
import pandas as pd

import time

from IPython.display import display
```

```python
sr = QueueRunner ()

queue_times_df = pd.DataFrame ()

queue_times_df["RunA_List"] = [sr.RunQueueList(1000, 1000, QueueList())]
queue_times_df["RunA_LinkedList"] = [sr.RunQueueList ( 1000, 1000, QueueLinkedList () )]

queue_times_df["RunB_List"] = [sr.RunQueueList(1, 1000, QueueList())]
queue_times_df["RunB_LinkedList"] = [sr.RunQueueList ( 1, 1000, QueueLinkedList () )]

queue_times_df["RunC_List"] = [sr.RunQueueList(1000, 1, QueueList())]
queue_times_df["RunC_LinkedList"] = [sr.RunQueueList ( 1000, 1, QueueLinkedList () )]


sr.plot_queue_comparisons_bg(queue_times_df)
display(queue_times_df)
```
