```python
from QueueRunner import QueueRunner
from StackRunner import StackRunner
from Queues import QueueLinkedList
from Queues import QueueList
from Stacks import StackList
from Stacks import StackLinkedList

import pandas as pd
from IPython.display import display
```

```python
sr = QueueRunner ()

queue_times_df = pd.DataFrame ()

queue_times_df["100-100_L"] = [sr.RunQueueList(100, 100, QueueList())]
queue_times_df["100-100_LL"] = [sr.RunQueueList ( 100, 100, QueueLinkedList () )]

queue_times_df["1-10000_L"] = [sr.RunQueueList(1, 10000, QueueList())]
queue_times_df["1-10000_LL"] = [sr.RunQueueList ( 1, 10000, QueueLinkedList () )]

queue_times_df["10000-1_L"] = [sr.RunQueueList(10000, 1, QueueList())]
queue_times_df["10000-1_LL"] = [sr.RunQueueList ( 10000, 1, QueueLinkedList () )]


sr.plot_queue_comparisons_bg(queue_times_df)
display(queue_times_df)
```
